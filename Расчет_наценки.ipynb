```python
import pandas as pd
```

```python
# читаю датафрейм только с нужными столбцами
df = pd.read_excel('/content/Тестовое.xlsx', sheet_name = 2, usecols=[0, 1, 3, 4, 7, 8])
df.head()
```

```python
# проверяю типы данных
df.info()
```

```python
# обрезаю датафрейм по дате и сумам продаж
df = df.drop(df[df['ДатаПродажи'] < '2023-04-01'].index)
df = df.drop(df[df['ДатаПродажи'] > '2023-04-10'].index)
df = df.drop(df[df['ОбщаяСумма'] <= 0].index)
df = df.drop(df[df['Себестоимость'] <= 0].index)
df.head()
```

```python
#  проверяю изменения
df.describe()
```

```python
# создаю колонку с значениями наценки
df['Наценка'] = df.apply(lambda x: x['ОбщаяСумма'] - x['Себестоимость'], axis=1)
df.head()
```

```python
# создаю категориальную колонку 1 - золото, 0 - серебро
df['Металл'] = df['ТоварнаяГруппа'].apply(lambda x: 0 if 'СИ' in x
                                          else 1)
df.head()
```

```python
# группирую по категории и вывожу средние значения наценки
df_group = df.groupby(['Металл']).agg({'Наценка': 'mean'})
df_group
```

Средняя наценка для серебра: 1294р
Средняя наценка для золота: 8634р
